```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
import warnings
from statsmodels.stats.diagnostic import acorr_ljungbox
```

```python
df = pd.read_csv('temperatura.csv')
df.head()
```

```python
# Plotando as s√©ries 'Temp_C' e 'Dew Point Temp_C'
plt.figure(figsize=(12, 5))
plt.plot(df['Temp_C'], label='Temp_C', color="deeppink")
plt.title('Temp_C - Time Series Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
# Teste ADF para 'Temp_C'

adf_test_temp = adfuller(df['Temp_C'])
print(f'ADF Statistic (Temp_C): {adf_test_temp[0]}')
print(f'p-value (Temp_C): {adf_test_temp[1]}')
```

```python
# ACF e PACF para 'Temp_C'

fig, axes = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(df['Temp_C'], ax=axes[0])
plot_pacf(df['Temp_C'], ax=axes[1])
plt.show()

df['Temp_C_diff'] = df['Temp_C'].diff()

plt.figure(figsize=(12, 5))
plt.plot(df['Temp_C_diff'], label='Temp_C_diff', color="deeppink")
plt.title('Temp_C_diff - Time Series Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()


adf_test_temp_diff = adfuller(df['Temp_C_diff'].dropna())
print(f'ADF Statistic (Differenced Temp_C): {adf_test_temp_diff[0]}')
print(f'p-value (Differenced Temp_C): {adf_test_temp_diff[1]}')
```

```python
# ACF e PACF para 'Temp_C_diff'
fig, axes = plt.subplots(2, 1, figsize=(10, 8))
plot_acf(df['Temp_C_diff'].dropna(), ax=axes[0])
plot_pacf(df['Temp_C_diff'].dropna(), ax=axes[1])
plt.show()
```

```python
# Configura√ß√£o para previs√£o com 'Temp_C'
h = 3
train_temp = df["Temp_C"].iloc[:-h]
test_temp = df["Temp_C"].iloc[-h:]
```

# Ajustando o modelo ARIMA para 'Temp_C'

```python

warnings.filterwarnings("ignore")

melhor_bic_temp = float("inf")
melhor_ordem_temp = None
melhor_modelo_temp = None

for p in range(0, 5):
    for d in range(0, 3):
        for q in range(0, 5):
            if p == 0 and d == 0 and q == 0:
                continue
            modelo_temp = ARIMA(train_temp, order=(p, d, q)).fit()
            bic_for_temp = float(modelo_temp.bic)
            if melhor_ordem_temp is None:
                melhor_bic_temp = bic_for_temp
                melhor_ordem_temp = (p, d, q)
                melhor_modelo_temp = modelo_temp
            else:
                if bic_for_temp < melhor_bic_temp:
                    melhor_bic_temp = bic_for_temp
                    melhor_ordem_temp = (p, d, q)
                    melhor_modelo_temp = modelo_temp

print(f"Melhor ARIMA para Temp_C {melhor_ordem_temp} BIC={melhor_bic_temp:.2f}")
```

```python
# Previs√£o para 'Temp_C'

residuals_temp = melhor_modelo_temp.resid

plt.figure(figsize=(12, 5))
plt.plot(residuals_temp)
plt.title('Res√≠duos do Modelo ARIMA para Temp_C')
plt.xlabel('Index')
plt.ylabel('Res√≠duo')
plt.show()

ljung_box_test_temp = acorr_ljungbox(residuals_temp, lags=[10], return_df=True)
print(ljung_box_test_temp)
```

```python
fc_temp = melhor_modelo_temp.get_forecast(steps=h)
men_temp = fc_temp.predicted_mean
ci_temp = fc_temp.conf_int(alpha=0.05)

fitted_temp = melhor_modelo_temp.fittedvalues
x_ci_temp = df.index[-h:]
plt.figure(figsize=(12,5))
plt.plot(df.index, df["Temp_C"].values, label="real")
plt.plot(fitted_temp.index, fitted_temp.values, label="fitted")
plt.plot(men_temp.index, men_temp.values, label="forecast")
plt.fill_between(x_ci_temp, ci_temp.iloc[:,0].values, ci_temp.iloc[:,1].values, alpha=0.2, label="IC95%")
plt.axvline(train_temp.index[-1], ls="--", label="split")
plt.legend()
plt.tight_layout()
plt.show()

print(fc_temp.summary_frame())
```

```python
# Reaplicando para 'Dew Point Temp_C'
# Plotando a s√©rie 'Dew Point Temp_C'
plt.figure(figsize=(12, 5))
plt.plot(df['Dew Point Temp_C'], label='Dew Point Temp_C', color="orange")
plt.title('Dew Point Temp_C - Time Series Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
# Teste ADF para 'Dew Point Temp_C'
adf_test_dew = adfuller(df['Dew Point Temp_C'])
print(f'ADF Statistic (Dew Point Temp_C): {adf_test_dew[0]}')
print(f'p-value (Dew Point Temp_C): {adf_test_dew[1]}')

# ACF e PACF para 'Dew Point Temp_C'
fig, axes = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(df['Dew Point Temp_C'], ax=axes[0])
plot_pacf(df['Dew Point Temp_C'], ax=axes[1])
plt.show()
```

```python
df['Dew Point Temp_C_diff'] = df['Dew Point Temp_C'].diff()

plt.figure(figsize=(12, 5))
plt.plot(df['Dew Point Temp_C_diff'], label='Dew Point Temp_C_diff', color="orange")
plt.title('Dew Point Temp_C_diff - Time Series Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()

adf_test_dew_diff = adfuller(df['Dew Point Temp_C_diff'].dropna())
print(f'ADF Statistic (Differenced Dew Point Temp_C): {adf_test_dew_diff[0]}')
print(f'p-value (Differenced Dew Point Temp_C): {adf_test_dew_diff[1]}')
```

```python
# ACF e PACF para 'Dew Point Temp_C_diff'
fig, axes = plt.subplots(2, 1, figsize=(10, 8))
plot_acf(df['Dew Point Temp_C_diff'].dropna(), ax=axes[0])
plot_pacf(df['Dew Point Temp_C_diff'].dropna(), ax=axes[1])
plt.show()

# Configura√ß√£o para previs√£o com 'Dew Point Temp_C'
train_dew = df["Dew Point Temp_C"].iloc[:-h]
test_dew = df["Dew Point Temp_C"].iloc[-h:]
```
